```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brownian_motion_st(s0, r, sigma, t):
  return (
      s0 * np.exp(
              (r - sigma**2 / 2) * t 
              + sigma * np.random.normal(0, t)
              )
  )

def mean_st(s0, r, T):
  return (
      s0 * np.exp(r * T)
  )
```

```python
s0 = 1
r = 0.2
T = 3
N = [10 ** i for i in range(1, 8)]
sigma = 0.1

st_mean = mean_st(s0, r, T)

d = []

for n in N:
  si = s0
  period = T / n
  for i in range(n):
    si = brownian_motion_st(si, r, sigma, period)
  d.append(np.abs(si - st_mean))
```

```python
fig, axes = plt.subplots()
axes.bar([str(i) for i in N], d)
axes.set_ylabel('|s(T) - E s(T)|')
axes.set_xlabel('N')
plt.show()
```
